```python
import pandas as pd
import os
import sys
from matplotlib import pyplot as plt
from datetime import datetime
os.chdir('E:\JUPYTER_files')
os.path.dirname(sys.executable)
```

```python
# input name of file you wish to read

csv = 'history 01 March 2023 - 00;36;39 - 4digit'
history = pd.read_csv(csv + '.csv')
```

```python
# change nothing here

writting = False
write_agg = False
logname = ''
agg = ''
for x in csv:
    if x == '-':
        if writting:
            write_agg = True
        writting = not writting
    if write_agg:
        agg += x
    if writting:
        logname += x
        
metriclist = []
metriclist.append(['accuracy','Accuracy'])
metriclist.append(['loss','BCE Loss'])
metriclist.append(['hamming_loss','Hammming Loss'])
metriclist.append(['precision','Precision'])
metriclist.append(['recall','Recall'])
metriclist.append(['auc','AUC'])
```

```python
# change only the aggregation variable

aggregation = 'Chapter'
for metric in metriclist:    
    title = aggregation + ' ' + metric[1]
    lbl = metric[0]

    if lbl not in ['loss','accuracy']:
        for x in history:
            if lbl in x:
                if "val_"in x:
                    temp_val = x
                else:
                    temp = x
    else:
        temp = lbl
        temp_val = 'val_'+lbl

    plt.plot(history[temp])
    plt.plot(history[temp_val])
    plt.title(title)
    plt.ylabel(lbl)
    plt.xlabel('epoch')
    plt.legend(['training', 'validation'], loc='lower right')
    figname = 'plot ' + agg + ' - ' + lbl + ' ' + logname + '.png'
    plt.savefig('logs/thesis_plots/' + figname, dpi=200)
    plt.show()
    print("Saved as: " + figname)
```
